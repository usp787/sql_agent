```python
import ollama
response = ollama.chat(
model='qwen2.5-coder:7b', 
messages=[
  {
    'role': 'user',
    'content': 'Write a Python function that returns the sum of even numbers in a list.',
  },
],
options={
        'num_ctx': 1024,
        'temperature': 0.2,
        'top_p': 0.9,
    }
)
print(response['message']['content'])
```

```python
def sum_of_evens(numbers):
    # Initialize the sum to zero
    total_sum = 0
    
    # Iterate through each number in the list
    for num in numbers:
        # Check if the number is even
        if num % 2 == 0:
            # Add the even number to the total sum
            total_sum += num
    
    # Return the total sum of even numbers
    return total_sum

numbers = [1, 2, 3, 4, 5, 6]
print(sum_of_evens(numbers))
```
